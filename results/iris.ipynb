```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 100)
```

```python
iris_results = pd.read_csv('../data/results/iris_results.csv')
```

```python
iris_results.columns
```

```python
means_columns = ['Unreduced_acc','Reduced_acc', 'Clf_unreducted_time', 'Clf_reducted_time','Reduction_time']
```

```python
def str_to_arr(s: str) -> str:
    s = s[1:-1]
    return np.fromstring(s, sep=',')
```

```python
str_to_arr(iris_results['Unreduced_acc'][0])
```

```python
for column in means_columns:
    iris_results[f'{column}_mean'] = iris_results[column].apply(lambda x: str_to_arr(x).mean())
```

```python
iris_to_show = iris_results.drop(columns=means_columns, axis=1)
```
