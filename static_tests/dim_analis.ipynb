```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# cv = 5
dims = [
    [1, 2, 3],
    [2, 5, 10, 20, 50],
    [2, 4, 5, 7, 8, 10],
    [2, 5, 10, 20, 50],
    [2, 4, 5, 7, 8, 10],
    [2, 5, 10, 20, 50],
    [2, 5, 10, 20, 50],
    [2, 4, 5, 7, 8, 10],
    [2, 4, 5, 7, 8, 10],
    [2, 4, 5, 7, 8, 10],
    [2, 4, 5, 7, 8, 10],
    [2, 4, 5, 7, 8, 10],
    [2, 5, 10, 20, 50],
    [2, 5, 10, 20, 50],
    [2, 5, 10, 20, 50],
    [2, 5, 10, 20, 50],
    [1, 2, 3, 4, 5],
    [2, 5, 10, 20, 50],
    [2, 5, 10, 20, 50],
    [1, 2, 3, 4, 5]
]
```

```python
sets_name = []
values_acc = []
values_times = []
avg_acc = []
avg_times = []
```

```python
with open('dims_results.txt') as f:
    lines = [line.rstrip() for line in f]
```

```python
def str_to_arr(s: str) -> str:
    s = s[1:-1]
    return np.fromstring(s, sep=',')
```

```python
i = 0
for line in lines:
    if i == 0:
        sets_name.append(line)
    elif i == 1:
        avg_acc.append(float(line.split(' ', 1)[0]))
        values_acc.append(str_to_arr(line.split(' ', 1)[1]))
    elif i == 2:
        values_times.append(str_to_arr(line.split(' ', 1)[1]))
        avg_times.append(float(line.split(' ', 1)[0]))
    if '---' in line:
        i = 0
    else:
        i += 1
```

```python
for set_name, acc, dim in zip(sets_name, values_acc, dims):
    plt.figure(figsize=(20, 5))
    plt.plot(dim, acc, 'x-')
    plt.plot([dim[0], dim[-1]], [1, 1], 'r-')
    plt.title(set_name)
    plt.ylabel('Reduction acc/Unreduction acc')
    plt.xlabel('Num of dims')
    plt.show()
```

```python
for set_name, acc, time in zip(sets_name, values_times, dims):
    plt.figure(figsize=(20, 5))
    plt.plot(time, acc, 'x-')
    plt.plot([time[0], time[-1]], [1, 1], 'r-')
    plt.title(set_name)
    plt.ylabel('Reduction time/Unreduction time')
    plt.xlabel('Time')
    plt.show()
```

```python
values_acc_dim_2 = []
for values in values_acc:
    values_acc_dim_2.append(values[0])
fig, ax = plt.subplots(1,1) 
fig.set_figheight(7)
fig.set_figwidth(23)
x = np.arange(0,20,1)
ax.plot(sets_name,values_acc_dim_2, 'x-')
ax.plot([0, 20], [1, 1], 'r-')
plt.title('Average acc for all sets for dim=2')
ax.set_ylabel('Reduction acc/Unreduction acc')
ax.set_xlabel('Sets')
ax.set_xticks(x)
ax.set_xticklabels(sets_name, rotation='vertical', fontsize=18)
plt.show()
```

```python
values_times_dim_2 = []
for times in values_times:
    values_times_dim_2.append(times[0])
fig, ax = plt.subplots(1,1) 
fig.set_figheight(7)
fig.set_figwidth(23)
x = np.arange(0,20,1)
ax.plot(sets_name,values_times_dim_2, 'x-')
ax.plot([0, 20], [1, 1], 'r-')
plt.title('Average time for all sets for dim=2')
ax.set_ylabel('Reduction time/Unreduction time')
ax.set_xlabel('Sets')
ax.set_xticks(x)
ax.set_xticklabels(sets_name, rotation='vertical', fontsize=18)
plt.show()
```
