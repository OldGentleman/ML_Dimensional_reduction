```python
import os
import pandas as pd
import numpy as np
from skimage.io import imread
from skimage.transform import resize

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from typing import Tuple

from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MaxAbsScaler


from sklearn.neighbors import KNeighborsClassifier

from zipfile import ZipFile


def Gems_load(img_size: Tuple[int, int]):
    gems_class = []
    gems_image = []
    #base_path = '../data/00_038.png'
 

    with ZipFile('../data/raw_data/d17 - gemstones_I.zip', 'r') as zip:
        zip.extractall('../data/preprocessed_data/d17 - gemstones_I')
    
    
    for root, dirs, files in os.walk('../data/preprocessed_data/d17 - gemstones_I/train'):
        path = root.split(os.sep)
        class_name = os.path.basename(root)# (len(path) - 1) * ''#,      
        for file in files:
            image = imread(root + '/' + file, as_gray=True)
                        
            if image.shape != img_size:
                print(image.shape)
                image = resize(image, img_size, anti_aliasing=True)
                print(image.shape)

            gems_image.append(image.ravel())
            gems_class.append(class_name)

    
    image_matrix = np.stack(gems_image)
    del gems_image
    
    le = LabelEncoder()
    le.fit(gems_class)
    gems_class = le.transform(gems_class)
    
    #print(image_matrix.shape)
    #print(gems_class_np.shape)
    
    print(image_matrix)
    print(gems_class)

    return gems_class, image_matrix


traget_gems, gems_images = Gems_load((50,50))

```
