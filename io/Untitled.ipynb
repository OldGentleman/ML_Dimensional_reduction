```python
import os
import pandas as pd
import numpy as np
from skimage.io import imread
from skimage.transform import resize

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from typing import Tuple


from sklearn.neighbors import KNeighborsClassifier

from zipfile import ZipFile


def Gems_load(img_size: Tuple[int, int]):
    gems_class = []
    gems_image = []


    with ZipFile('../data/raw_data/d17 - gemstones_I.zip', 'r') as zip:
        zip.extractall('../data/preprocessed_data/d17 - gemstones_I')

    for root, dirs, files in os.walk('../data/preprocessed_data/d17 - gemstones_I/train'):
        path = root.split(os.sep)
        class_name = os.path.basename(root)# (len(path) - 1) * ''#,      
        for file in files:
            image = imread(root + '/' + file)

            if image.shape[0] != img_size or image.shape[1] != img_size:
                        image = resize(image, img_size, anti_aliasing=True)

            gems_image.append(image.ravel())
            gems_class.append(class_name)


    
    gems_class_np = np.array(gems_class, dtype=object)#.reshape(-1,1)
    gems_image_np = np.array(gems_image, dtype=object).reshape(-1,1)
    
    #print(gems_image)
    
    return gems_class_np, gems_image_np


traget_gems, gems_images = Gems_load((128,128))

```
